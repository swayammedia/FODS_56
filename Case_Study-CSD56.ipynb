```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv('ecommerce_dataset_1.csv')
df1=pd.read_csv('ecommerce_dataset_2.csv')
df1=pd.DataFrame(df)
df2=pd.DataFrame(df1)
df_merged=pd.merge(df1,df2,on='ProductID')
```

```python
df_merged
```

```python
df_merged.shape
```

```python
df_merged.drop_duplicates(inplace=True)
```

```python
q1=df_merged['OrderPrice'].quantile(0.25)
q1
```

```python
q3=df_merged['OrderPrice'].quantile(0.75)
q3
```

```python
IQR=q3-q1
```

```python
lb=q1 - 1.5 * IQR
ub=q3 + 1.5 * IQR
```

```python
outlier=((df_merged[['OrderPrice']]<lb) | (df_merged[['OrderPrice']]>ub)).any(axis=1)
df_filtered=df_merged[~outlier]
df_filtered
```

```python
scaler=MinMaxScaler()
df_merged['Quantity_Normalized']=scaler.fit_transform(df_merged[['OrderQuantity']])
df_merged['Price_Normalized']=scaler.fit_transform(df_merged[['OrderPrice']])
df_merged
```

```python
df_encoded=pd.get_dummies(df_merged,columns=['Category'])
df_encoded
```
